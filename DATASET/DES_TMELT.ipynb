```python
#devo trasformare i t melt di df1 in kelvin
df1["Tmelt_HBA"] = df1["Tmelt_HBA"] + 273.15
df1["Tmelt_HBD"] = df1["Tmelt_HBD"] + 273.15
df1["Mixed_Tmelt "] = df1["Mixed_Tmelt "] + 273.15
df1.head()
```

```python
#aggiungo le colomnne a df1
df1["N°C"] = 2
df1["DES_T"] = np.nan
df1["NAME_HBA"] = np.nan
df1["NAME_HBD"] = np.nan
df1["REF#DOI"] = np.nan
df1.head()
```

```python
#droppiamo le colonne a df2
df2 = df2.drop(columns=["Phase diagram (Yes/No)"])
df2.head()
```

```python
#ordiniamo i dati delle due tabelle
df1 = df1[["N°C", "DES_T", "NAME_HBA", "SMILES_HBA", "Tmelt_HBA", "NAME_HBD", "SMILES_HBD", "Tmelt_HBD", "X_HBA", "X_HBD", "Mixed_Tmelt ", "REF#DOI"]]	               
df1.head()
```

```python
df2 = df2[["Number of components", "Type of DES", "Component#1", "Smiles#1", "T#1", "Component#2", "Smiles#2", "T#2", "X#1 (molar fraction)", "X#2 (molar fraction)", "Tmelt, K", "Reference (DOI)" ]]
df2.head()
```

```python
#rinominiamo le colonne
df2 = df2.rename(
    columns={
    'Number of components': 'N°C',
    "Type of DES": "DES_T",
     "Component#1":   "NAME_HBA",
    'Smiles#1': 'SMILES_HBA',
    'T#1': 'Tmelt_HBA',
    "Component#2":   "NAME_HBD",
    'Smiles#2' : 'SMILES_HBD',
    'T#2': 'Tmelt_HBD',
    'X#1 (molar fraction)': 'X_HBA',
    'X#2 (molar fraction)': 'X_HBD',
    "Tmelt, K" : "Mixed_Tmelt ",
    "Reference (DOI)" : "REF#DOI"
    })

df2.head()
```

```python
#guardiamo i dat comuni tra i due dataset
df_comuni = pd.merge(df1, df2, on=["SMILES_HBA", "SMILES_HBD", "X_HBA"])
df_comuni
#il dato numero 2 è anomalo: temperature di fusione diverse èer lo stesso DES
#gli altri 8 dati in comune sono simili (+-10K) e terremo entrambe le copie per non perdere nessuna informazione
```
